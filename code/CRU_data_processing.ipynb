```python
# reshape to get the yearly averages
temp_reshaped = temp.reshape(-1, 12, 360, 720).mean(axis=1)
# reshape to make each latitude, longitude pair it's own row
temp_reshaped = temp_reshaped.reshape(-1, 360, 720)
```

```python
temp_reshaped.shape
```

```python
# Assuming you have a list of years (replace with actual years)
years = [year for year in range(1901, 1901 + temp_reshaped.shape[0])]
```

```python
years
```

```python
temp_2d_array = temp_reshaped.reshape(temp_reshaped.shape[0], -1)
```

```python
# Create a DataFrame
df = pd.DataFrame(temp_2d_array, index=years)

# Transpose the DataFrame to have latitude, longitude pairs as columns
df = df.transpose()

# Optional: Add latitude and longitude as MultiIndex
df['Latitude'] = np.tile(np.arange(-90, 90, 0.5), temp_reshaped.shape[2])
df['Longitude'] = np.repeat(np.arange(0, 360, 0.5), temp_reshaped.shape[1])

# Set the index to the 'Latitude' and 'Longitude' columns
df.set_index(['Latitude', 'Longitude'], inplace=True)
```

```python
df.dropna(inplace=True)
df
```

```python
import geopandas as gpd
```

```python
shapefile_path = '../data/CRU/natural_earth_vector/10m_cultural/ne_10m_admin_0_countries.shp'
```

```python
gdf_countries = gpd.read_file(shapefile_path)
print(gdf_countries.head())
```

```python
# Reshape the DataFrame to have latitude and longitude as columns
df = df.transpose().reset_index()
df.columns = ['Latitude', 'Longitude'] + [f'Year_{year}' for year in df.columns[2:]]

# Merge the temperature DataFrame with the country geometries based on latitude and longitude
gdf_temp = gpd.GeoDataFrame(pd.merge(gdf_countries, df, how='inner', left_on=['Latitude', 'Longitude'], right_on=['Latitude', 'Longitude']))

# Group by country and calculate the mean temperature for each country
gdf_temp_grouped = gdf_temp.groupby('ADMIN').mean()

# Reset the index
gdf_temp_grouped.reset_index(inplace=True)
```
