```python
filename = "../data/CRU/cru_ts4.07.1901.2022.tmp.dat.nc"
data = netCDF4.Dataset(filename)
```

```python
print(data.variables.keys())
```

```python
temp = data.variables['tmp'][:]
lon = data.variables['lon'][:]
lat = data.variables['lat'][:]
time = data.variables['time']
```

```python
# reshape to get the yearly averages
temp_reshaped = temp.reshape(-1, 12, 360, 720).mean(axis=1)
# reshape to make each latitude, longitude pair it's own row
temp_reshaped = temp_reshaped.reshape(-1, 360, 720)
```

```python
temp_reshaped.shape
```

```python
# Assuming you have a list of years (replace with actual years)
years = [year for year in range(1901, 1901 + temp_reshaped.shape[0])]
```

```python
years
```

```python
temp_2d_array = temp_reshaped.reshape(temp_reshaped.shape[0], -1)
```

```python
# Create a DataFrame
df = pd.DataFrame(temp_2d_array, index=years)

# Transpose the DataFrame to have latitude, longitude pairs as columns
df = df.transpose()

latitude_index = np.arange(-90, 90, 0.5)
longitude_index = np.arange(0, 360, 0.5)

df.index = pd.MultiIndex.from_product([latitude_index, longitude_index], names=['Latitude', 'Longitude'])
df.dropna(inplace=True)
# Reset index to make 'Latitude' and 'Longitude' regular columns
df.reset_index(inplace=True)

# # Step 4: Aggregate by larger squares (5 by 5)
# latitude_bins = np.arange(-90, 91, 5)
# longitude_bins = np.arange(0, 361, 5)

# Group by latitude and longitude bins, take the mean, and ignore missing values
# df_aggregated = df.groupby([pd.cut(df['Latitude'], bins=latitude_bins),
#                             pd.cut(df['Longitude'], bins=longitude_bins)]).mean()

# # Extract the center coordinates of each bin and add them as columns
# df_aggregated['Latitude_Center'] = df_aggregated['Latitude'].mid
# df_aggregated['Longitude_Center'] = df_aggregated['Longitude'].mid
```
